# Another ways Data set with multiple parameters

```python
days=[54,76,86,98,4,5]
travel=[45,32,54,87,43,2]
food=[65,23,45,23,87,5]
expenses=[56,76,34,23,54,12]
plt.xlabel('Days')
plt.ylabel("Travel")  # use for giving labels to the graph
plt.title('Place')
plt.plot(days,travel,food,expenses , linewidth=5, linestyle="dotted")
```

```python
plt.plot(days,travel,food,expenses , linewidth=5, marker="+" ,linestyle="" ,markersize="30") 
# new properties were added like marker /symbol amd marker size
```

```python
plt.plot(days,travel,food,expenses , linewidth=5, marker="P" ,linestyle="" ,markersize="30") 
```

## Now we will run  how to make bar graphs using matplotlib


```python
# kisi bhi manner mai ham files import kar skte hai kya 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
# import pandas as pd
# %matplotlib inline
# import matplotlib.pyplot as plt
# import numpy as np
```

```python
Technologies=['Data science','Web-Deve ','Flutter-Dev','Cloud Computing'] # x  represent Technologies
Salary=[45000,35000,26076,40000]  #y represent salary
Developers=[4000,7600,3400,1200]
```

```python
xpos = np.arange(len(Technologies))  # it shows the value in the given set
xpos
```

```python
plt.bar(Technologies,Salary)
```

```python
plt.xticks(xpos,Technologies)
plt.ylabel("Salary(Average)")  # shows labeling, xticks will plot technology rates

plt.bar(xpos,Salary, label="Salary")

plt.title('World of Technology ')   # we have given labeling to it

plt.legend()
```

```python
plt.bar(xpos-0.2,Salary, width=0.4 ,label="Salary")
plt.bar(xpos+0.2,Developers, width=0.4 ,label="Developers")

plt.xticks(xpos,Developers)
plt.ylabel("Salary(Average)")
plt.title('world of technologies')
plt.legend()  # use to give specific area to the paramenters 
```
